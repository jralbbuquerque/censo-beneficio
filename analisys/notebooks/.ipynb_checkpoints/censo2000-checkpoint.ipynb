## **1. Indíce de Gini**
* Indicador que mede a desigualdade de renda
* G = 1 corresponde a total desigualdade e G = 0 corresponde a total igualdade de renda
* Pode ser cálculado como:

$G = 1 - \sum_{i=1}^{k=n-1} (p_{k+1} - p_{k})(r_{k+1} + r_{k})$
* Onde: 
    - G - Coeficiente de Gini
    - P - Proporção acumulada da "população"
    - R - Proporção acumulada de "renda"
    
* Obs: é necessário organizar **R** de forma crescente: 

$r_{1} ≤ r_{2} ≤ r_{3} ... ≤ r_{i} ≤ ... ≤ r_{n}$

````python
# ---------- ORGANIZAÇÃO DOS DADOS DE ENTRADA ---------- #

def dadosEntrada( sql ):
    # sql = "select coalesce(v7616,0) as renda, round(p001) as peso from censo_2000.domicilios_pa where v0103=1502400"
 
    df = pd.read_sql_query( sql, conn )     # Realiza a leitura do sql e atrabui o resultado a DataFrame
    rendimento = []                         # Cria um array rendimento vazio
    renda = df[ 'renda' ]                   # Extrai apenas a coluna renda
    peso  = round( df[ 'peso' ] )           # Extrai apenas a coluna penso e arrendonda seu valor

    for i in range( len( renda ) ):
        for j in range( int( peso[ i ] ) ):   
            rendimento.append( renda[ i ] ) # Realiza a expansão da base comprimida 

    renda = np.sort( rendimento )           # Renda em ordem crescente 
    
    return renda

# retorna como saída os dados renda ordenados 
````

````python
# ---------- CALCULO DO ÍNDICE DE GINI ---------- #

def gini( sql ):
    inicio = time.time() # Inicialização do tempo de processamento
    
    renda = dadosEntrada( sql )
    
    n = len( renda )      # Tamanho da amostra de dados
    m = sum( renda ) / n  # Renda média

    p = [ 0 ]  # Inicializa o array da proporção acumulada da população
    r = [ 0 ]  # Inicializa o array da proporção acumulada de renda

    for i in range( 1, ( n + 1 ) ):
        p.append( i / n )                           # Proporção acumulada da população
        r.append( sum( renda[ 0:i ] ) / ( n * m ) ) # Proporção acumulada da renda

    gini_parcial = ( renda[ 0 ] / ( n * m ) ) * p[ 0 ]  # Cálculo do indíce de Gini para a primeira componente de renda

    aux = 0 # Variável auxiliar utilizada no cálculo do íncide de Gini

    for i in range( 0, n ):
        aux = ( p[ i + 1 ] - p[ i ] ) * ( r[ i + 1 ] + r[ i ] ) # Fórumula do indíce de Gini
        gini_parcial = gini_parcial + aux 

    gini = 1 - gini_parcial # Cálculo final

    fim = time.time() # Finalização do tempo de processamento
    tempo = fim - inicio
    
    return gini, tempo, p, r

#retorna como saída: o índice de gini, tempo de processamento, pop acumulada e renda acumulada (utilizados nos plots)
````

````python
# ---------- PLOT DA CURVA DE LORENZ---------- #

def plotagem( p, r ):
    plt.plot( p, r, 'k' )
    plt.plot( [ 0, 1 ], 'r' )
    plt.ylim( 0, 1 ), plt.xlim( 0, 1 )
    plt.fill_between( p, r, facecolor = 'y' )

    plt.xlabel( 'Proporção de população' )
    plt.ylabel( 'Proporção de renda' )
    plt.title( 'Curva de Lorenz' )
    plt.show()
````

````python
# ---------- CALCULO DO ÍNDICE DE GINI PARA RENDA DOMICILIAR ---------- #

sql = "SELECT DISTINCT(v0103) AS cidade FROM censo_2000.domicilios_br"
cidades = pd.read_sql_query( sql, conn )

g_total = []

for i in range( 0, len( cidades ) ):
    
    sql = "SELECT \
            COALESCE(v7616,0) AS renda, ROUND(p001) AS peso \
            FROM censo_2000.domicilios_br \
            WHERE v0103 = %d" %cidades.loc[ i, 'cidade' ]
    
    sql_renda = "SELECT \
                    SUM( v7616 * p001 ) / SUM( p001 ) AS RENDA_v7616 \
                    FROM censo_2000.domicilios_br \
                    WHERE v0103 = %d and v7616 IS NOT NULL"  %cidades.loc[ i, 'cidade' ]
    
    df_renda = pd.read_sql_query( sql_renda, conn )
    
    # Retornando um array com [índice de gini], [tempo de execução], ...
    # proporção acumulado da população] e [proporção acumulada de renda]
    g = gini( sql ) 
    
    g_total.append( [ cidades.loc[ i, 'cidade' ], g[0], 
                     df_renda.loc[ 0, 'renda_v7616' ] ] )

    
gini_domicilio = pd.DataFrame( g_total, columns = 'Cod Gini RendaMediaDomiciliar'.split() )
gini_domicilio.to_csv( 'domicilios_pa.csv', sep = '\t' )

gini_domicilio
````

````python
sql = "SELECT DISTINCT(v1103) AS cidade FROM censo_2000.pessoas_pa"
cidades_pa = pd.read_sql_query( sql, conn )

g_total = []

for i in range( 0, len( cidades_pa ) ):
    
    sql = "SELECT \
            COALESCE(v4512,0) AS renda, ROUND(p001) AS peso \
            FROM censo_2000.pessoas_pa \
            WHERE v1103 = %d AND v4512 is not null" %cidades_pa.loc[ i, 'cidade' ]
    
    sql_renda_4512 = "SELECT \
                        SUM( v4512 * p001 ) / SUM( p001 ) AS RENDA_v4512 \
                        FROM censo_2000.pessoas_pa \
                        WHERE v1103 = %d AND v4572 > 10"  %cidades_pa.loc[ i, 'cidade' ]

    sql_renda_4525 = "SELECT \
                        SUM( v4525 * p001 ) / SUM( p001 ) AS RENDA_v4525 \
                        FROM censo_2000.pessoas_pa \
                        WHERE v1103 = %d AND v4572 > 10"  %cidades_pa.loc[ i, 'cidade' ]
    
    sql_renda_4614 = "SELECT \
                        SUM( v4614 * p001 ) / SUM( p001 ) AS RENDA_v4614 \
                        FROM censo_2000.pessoas_pa \
                        WHERE v1103 = %d AND v4572 > 10"  %cidades_pa.loc[ i, 'cidade' ]
    
    sql_renda_benef = "SELECT \
                        SUM( v4573 * p001 ) / SUM( p001 ) AS RENDA_BENEFICIO \
                        FROM censo_2000.pessoas_pa \
                        WHERE v1103 = %d AND v4572 > 10"  %cidades_pa.loc[ i, 'cidade' ]
    
    
    sql_pop = "SELECT \
                    SUM( p001 ) AS POPULACAO \
                    FROM censo_2000.pessoas_pa \
                    WHERE v1103 = %d" %cidades_pa.loc[ i, 'cidade' ] 
    
    
    df_pop = pd.read_sql_query( sql_pop, conn )
    df_renda_4512 = pd.read_sql_query( sql_renda_4512, conn )
    df_renda_4522 = pd.read_sql_query( sql_renda_4525, conn )
    df_renda_4614 = pd.read_sql_query( sql_renda_4614, conn )
    df_renda_benef = pd.read_sql_query( sql_renda_benef, conn )
    
    # Retornando um array com [índice de gini], [tempo de execução], ...
    # proporção acumulado da população] e [proporção acumulada de renda]
    g = gini( sql ) 
    g_total.append( [ cidades_pa.loc[ i, 'cidade' ], 
                      g[0], 
                      df_pop.loc[ 0, 'populacao'],
                      df_renda_4512.loc[ 0, 'renda_v4512'],
                      df_renda_4522.loc[ 0, 'renda_v4525'],
                      df_renda_4614.loc[ 0, 'renda_v4614'],
                      df_renda_benef.loc[ 0, 'renda_beneficio'] ] )

gini_pop = pd.DataFrame( g_total, columns = 'Cod Gini Populacao RendaMedia4512 \
                                                RendaMedia4525 RendaMedia4614 \
                                                RendaMediaBeneficios'.split() )

gini_pop.to_csv( 'pessoas_pa.csv', sep = '\t' )
gini_pop
````
